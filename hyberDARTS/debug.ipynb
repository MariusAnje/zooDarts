```python
import torch
from matplotlib import pyplot as plt
```

```python
f = torch.load( "gradients.pt", map_location="cpu")
```

```python
c = f[0]
l = f[1]
```

```python
size = torch.Size([len(l), len(l[0]), len(l[0][0]), len(l[0][0][0])])
```

```python
lT = torch.zeros(size)
for i in range(size[0]):
    for j in range(size[1]):
        for k in range(size[2]):
            lT[i,j,k] = l[i][j][k]

cT = torch.zeros(size)
for i in range(size[0]):
    for j in range(size[1]):
        for k in range(size[2]):
            cT[i,j,k] = c[i][j][k]
```

```python
lT = lT.view(-1,6,3)
cT = cT.view(-1,6,3)
```

```python
plt.scatter(cT[1:,0,0], lT[1:,0,0], alpha = 0.5)
plt.axis([-0.04, 0.1, -0.0001, 0.00001])
```
